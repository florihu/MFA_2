First lets define the prior distributions
Properties of the Uniform Distribution
The uniform distribution has the following properties:

Mean: (a + b) / 2
Median: (a + b) / 2
Standard Deviation: √(b – a)2 / 12
Variance: (b – a)2 / 12

Initiate the Prior distributions


Initiate the prior distributions:

```python
import numpy as np
from scipy.stats import uniform
import seaborn as sns


y_0 = uniform.rvs(loc=50, scale = np.sqrt(100**2/12), size=1000)
n_eff = uniform.rvs(loc=.5, scale = np.sqrt(1**2/12), size=1000)

sns.histplot(n_eff)
```
